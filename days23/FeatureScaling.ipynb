```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['User ID', 'Gender'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.3, random_state=0)
```

```python
X_train.shape,X_test.shape
# X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
sc.fit(X_train)
X_train_scaled =sc.transform(X_train)
X_test_scaled  =sc.transform(X_test)
```

```python
X_train_scaled =pd.DataFrame(X_train_scaled,columns =X_train.columns)
X_test_scaled =pd.DataFrame(X_test_scaled,columns =X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig ,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'])
ax2.set_title("After Scaling")
plt.show()
```

```python
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Plot before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Age'], ax=ax1, label='Age')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1, label='Estimated Salary')
ax1.legend()

# Plot after scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Age'], ax=ax2, label='Age')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2, label='Estimated Salary')
ax2.legend()

plt.tight_layout()
plt.show()
```
